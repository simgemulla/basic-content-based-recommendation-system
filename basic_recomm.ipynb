```python
import pandas as pd
import numpy as np

credits = pd.read_csv("tmdb_5000_credits.csv")
movies = pd.read_csv("tmdb_5000_movies.csv")

credits.head()
```

```python
movies.head()
```

```python
print("movies shape: ", movies.shape)
print("credits shape: ", credits.shape)
```

```python
credits_renamed = credits.rename(index = str, columns = {"movie_id" : "id"})
movies_merged = movies.merge(credits_renamed, on = 'id')

movies_merged.head()
```

```python
movies_clean = movies_merged.drop(columns = ['homepage', 'title_x', 'title_y', 'status', 'production_countries'])

movies_clean.head()
```

```python
print(movies_clean.info())
```

```python
print(movies_clean.head(1)['overview'])
```

The aim here is to build a basic content based recommendation system. For example if the user gives us a name of a movie, we need to return a similar movie name.

```python
from sklearn.feature_extraction.text import TfidfVectorizer

tfv = TfidfVectorizer(min_df = 3, max_features = None, 
                     strip_accents = 'unicode', analyzer = 'word', token_pattern = r'\w{1,}',
                     ngram_range=(1, 3), stop_words = 'english')
```

```python
# Fitting the TF-IDF on the 'overview' text

tfv_matrix = tfv.fit_transform(movies_clean['overview'].values.astype('U'))
#x = v.fit_transform(df['Review'].values.astype('U'))  ## Even astype(str) would work
print(tfv_matrix)
print(tfv_matrix.shape)
```
